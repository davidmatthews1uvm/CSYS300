```python
import matplotlib.pyplot as plt
```

```python
color_invalid = "C3"
color_boy = "C0"
color_girl = "C4"
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:

WORLDS_PER_CHILD = 1
TOTAL_WORLDS = (WORLDS_PER_CHILD*2)**2 # each child could be boy or girl

invalid_worlds = (TOTAL_WORLDS - (WORLDS_PER_CHILD*4) + 1) # no girl born
one_boy_worlds = (WORLDS_PER_CHILD * 2)
two_girl_worlds = (WORLDS_PER_CHILD * 2-1) 

sizes = [d/TOTAL_WORLDS for d in [invalid_worlds, one_boy_worlds, two_girl_worlds]]
colors = [color_invalid, color_boy, color_girl]
labels = ["No girls born", "girl born, boy born", "two girls born"]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, startangle=90, autopct='%1.1f%%', colors=colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig("part9_1girl.pdf", bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:

WORLDS_PER_CHILD = 7
TOTAL_WORLDS = (WORLDS_PER_CHILD*2)**2 # each child could be boy or girl or born on any day.

invalid_worlds = (TOTAL_WORLDS - (WORLDS_PER_CHILD*4) + 1) # no girl born on Tuesday
one_boy_worlds = (WORLDS_PER_CHILD * 2)
two_girl_worlds = (WORLDS_PER_CHILD * 2-1) 

sizes = [d/TOTAL_WORLDS for d in [invalid_worlds, one_boy_worlds, two_girl_worlds]]
colors = [color_invalid, color_boy, color_girl]
fig, axs = plt.subplots(1,2)
ax1 = axs[0]
wedges, texts, autotexts = ax1.pie(sizes, explode=(0, 0.1, 0.1), startangle=90, autopct='%1.1f%%', colors=colors)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.legend(wedges, ["Impossible worlds", "Worlds with one girl", "Worlds with two girls"],
          title="Worlds",
          loc="best",
          bbox_to_anchor=(1, 0, 0.5, 1))

ax2 = axs[1]
wedges, texts, autotexts = ax2.pie([one_boy_worlds/(one_boy_worlds+two_girl_worlds), two_girl_worlds/(one_boy_worlds+two_girl_worlds)],
                                   startangle=90, autopct='%1.1f%%', colors=colors[1:])

ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("One Girl Born on Tuesday")
plt.savefig("part9_1girl_tuesday.pdf", bbox_inches='tight', pad_inches=0)
plt.show()
```
