```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
def get_count(s):
    if "1" in s:
        return 1
    if "2" in s:
        return 2
    if "4" in s:
        return 4
    if "8" in s:
        return 8

def parse_wiki_table(s, verbose=False):
    data = []
    for row in s.split("|-")[2:]:
        cols = row.split("\n")
        oarpeople = cols[1].strip()[1:]
        oarpeople = oarpeople[:oarpeople.find(" ")][1:]
        time = cols[2].strip()[1:]
        mins_split, seconds_split = time.split(":")
        seconds = float(mins_split) * 60 + float(seconds_split)
        data.append((get_count(oarpeople), seconds))
        if verbose:
            print(oarpeople, get_count(oarpeople), time, seconds)
    return data


oarpeople_m, times_m = zip(*parse_wiki_table(wiki_men_rowing_records))
oarpeople_w, times_w = zip(*parse_wiki_table(wiki_women_rowing_records))
df = pd.concat([pd.DataFrame({"oarpeople":oarpeople_m, "time":times_m, "sex":"m"}),
                pd.DataFrame({"oarpeople":oarpeople_w, "time":times_w, "sex":"w"})])
df["km/hr"] = 3600*2/df["time"]
```

```python
log_oarspeople = np.log10(df["oarpeople"])
log_speed = np.log10(df["km/hr"])

slope, intercept, r_value, p_value, std_err = stats.linregress(log_oarspeople, log_speed)
predict_log_speed = lambda n: intercept + slope * n

print("Slope: %.6f"%slope)
print("Intercept: %.6f"%intercept)
print("R: %.6f (R^2: %.6f), P_val: %.6E"%( r_value, r_value**2, p_value))

np.power(10, predict_log_speed(log_oarspeople))
```

```python
fig, ax = plt.subplots(1, figsize=(12,8))

ax.plot(df["oarpeople"], np.power(10, predict_log_speed(log_oarspeople)), c="C0")
ax.scatter(df["oarpeople"], df["km/hr"], c="C1")
ax.set_xlabel("Number of oarpeople", fontsize=18)
ax.set_ylabel("Speed (km/hr)", fontsize=18)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xticks([1,2,4,8])
ax.set_xticklabels(["1", "2", "4","8"])
ax.set_title("2000m Rowing records", fontsize=24)
ax.text(1.2, 22, r"$R^2$: %.6E"%(r_value**2), fontsize=18)
ax.text(1.2, 21, r"speed(n) = $n^{%.4f} + %.4f$"%(slope, np.power(10,intercept)), fontsize=18)
plt.savefig("part2.pdf", bbox_inches='tight', pad_inches=0)
```
